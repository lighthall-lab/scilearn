```python
import os
import pandas as pd
```

```python
source_dir=os.path.join('..','sourcedata','qualtrics')
derivs_dir=os.path.join('..','derivatives')
```

## Comprehension

###### Import Comprehension data

```python
frame=pd.DataFrame()
for file in os.listdir(source_dir):
    if file.startswith('Comprehension-'):
        df = pd.read_csv(os.path.join(source_dir,file))
        df['time']=os.path.basename(file)[14]
        frame=frame.append(df)
```

###### Filter & Sort

```python
frame=frame.sort_values(['SSID','time']).set_index(['SSID','time'])
frame.head()
```

###### Output

```python
frame.to_csv(os.path.join(derivs_dir,'comprehension_data.csv'))
```

# Survey Blocks

###### Import & combine Survey data

```python
frames=[]
for file in os.listdir(source_dir):
    if file.startswith('Survey-'):
        df = pd.read_csv(os.path.join(source_dir,file))
        frames.append(df)
frame = frames[1].merge(frames[0],'outer',on='SSID')
frame.head()
```

###### Output

```python
frame.to_csv(os.path.join(derivs_dir,'survey_data.csv'),index=False)
```
