```python
import os
import pandas as pd
```

```python
source_dir=os.path.join('..','sourcedata','qualtrics')
derivs_dir=os.path.join('..','derivatives')
```

## Comprehension

###### Import Comprehension data

```python
frame=pd.DataFrame()
for file in os.listdir(source_dir):
    if file.startswith('Comprehension-'):
        df = pd.read_csv(os.path.join(source_dir,file))
        df['time']=os.path.basename(file)[14]
        frame=frame.append(df)
```

###### Filter & Sort

```python
frame=frame[frame['Finished']=='1']
frame.head()
```

```python
frame.columns
```

```python
columns_to_drop=[
    'StartDate','EndDate','Status','IPAddress','Finished',
    'RecipientLastName','RecipientFirstName','RecipientEmail',
    'LocationLatitude','LocationLongitude','LocationAccuracy',
    'ResponseID','ResponseSet','ExternalDataReference',
    'Score-weightedAvg','Score-weightedStdDev'#,'RO-BR-FL_23',
    #'DO-Q-01a','DO-Q-02a', 'DO-Q-03a', 'DO-Q-04a',
    #'DO-Q-05a', 'DO-Q-06a', 'DO-Q-07a', 'DO-Q-08a',
    #'DO-Q-09b', 'DO-Q-10a', 'DO-Q-11b', 'DO-Q-12b',
    #'DO-Q-13b','DO-Q-14a', 'DO-Q-15b', 'DO-Q-17b'
]
columns=[
    'SSID','time','Score-sum',
    '01a','02a','03a','04a','05a','06a','07a','08a','09b',
    '10a','11b','12b','13b','14a','15b','16b','17b'
]
```

```python
frame=frame[columns]
frame.head()
```

```python
frame=frame.sort_values(['SSID','time']).set_index(['SSID','time'])
frame.head()
```

###### Output

```python
frame.to_csv(os.path.join(derivs_dir,'sub-all_task-comp_beh.csv'))
```
